```python
import kagglehub
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split

pd.set_option("display.max_columns",None)
pd.set_option("display.width",100)

# Download latest version
path = kagglehub.dataset_download("brendan45774/test-file")
print("Path to dataset files:", path)
filepath=f"{path}/tested.csv"
df=pd.read_csv(filepath)
print("Intial Dataframe (first 5 rows):")
print(df.head())


# y_target=df['Species']

print("\nExploring data")
df.info()

print("\ndescribe method")
print(df.describe())

print("\nmissing values")
print(df.isnull().sum())

print("\nDrop method")
df_processed=df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'],axis=1)
print(df_processed.head())

print("\nHandling missing values")
print(df_processed.isnull().sum())


age_median=df_processed['Age'].median()
df_processed['Age']=df_processed['Age'].fillna(age_median)
print(f"\nimputing age with median value:{age_median:.2f}")

fare_median=df_processed['Fare'].median()
df_processed['Fare']=df_processed['Fare'].fillna(fare_median)
print(f"imputing fare with median value:{fare_median:.2f}")


print("After handling missing values")
print(df_processed.isnull().sum())


print("\nHandling categorical data")
le=LabelEncoder()
print("\nBefore label encoding (for sex column)")
print(df_processed['Sex'].head)
print("\nfirst the encoder fit and tranform the data")
df_processed['Sex'] = le.fit_transform(df_processed['Sex'])
print(df_processed['Sex'].head)
print(f"classes found for 'sex':{le.classes_} ")

print("\nOne hot encoding('embark column)")
one=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
embark_column=df_processed[['Embarked']]
print("\nBefore")
print(df_processed[['Embarked']].head)


one_hot_column=one.fit_transform(embark_column)
onehot_df=pd.DataFrame(one_hot_column,columns=one.get_feature_names_out(['Embarked']))
onehot_encoded_column=one.fit_transform(df_processed[['Embarked']])
print(onehot_df.head())

df_processed=df_processed.drop('Embarked',axis=1)
df_processed=pd.concat([df_processed,onehot_df],axis=1)


print("\nAfter")
print(df_processed.head())

print("\nAppliying Standardization")
numerical_column=['Age','Fare','SibSp','Parch','Pclass']
scaler=StandardScaler()

print("Before scaling")
print(df_processed[numerical_column].head())
df_processed[numerical_column]=scaler.fit_transform(df_processed[numerical_column])
print("After scaling")
print(df_processed[numerical_column].head())

print("\nSplitting data")
x=df_processed.drop('Survived',axis=1)
y=df_processed['Survived']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
print("\nDataset has been split into training and testing sets.")
print(f"Shape of X_train: {x_train.shape}")
print(f"Shape of X_test: {x_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")

print("\nFirst 5 rows of the final training features (X_train):")
print(x_train.head())





















































```
